```python
from collections import defaultdict


# unique ages of each mouse by lab

age_by_lab = defaultdict(lambda: defaultdict(set))

for lab_id, grp in train_metadata.groupby("lab_id"):
    for col in cols:
        age_by_lab[lab_id][col].update(grp[col].dropna().unique())
        
nested_dict = {
    lab: {k: ', '.join(v) if v else '' for k, v in inner.items()}
    for lab, inner in age_by_lab.items()
}

# create dataframe
df = pd.DataFrame.from_dict(nested_dict, orient='index')

# display
df
```

```python
import pandas as pd
from collections import defaultdict

# Step 1: collect counts
strain_counts = defaultdict(lambda: defaultdict(int))

for lab_id, grp in train_metadata.groupby("lab_id"):
    for col in cols:
        for strain, count in grp[col].value_counts().items():
            if pd.notna(strain):
                strain_counts[lab_id][strain] += count

# Step 2: convert to DataFrame
df = pd.DataFrame.from_dict(
    {lab: dict(strains) for lab, strains in strain_counts.items()},
    orient='index'
).fillna(0).astype(int)

df.index.name = 'lab_id'
df.columns.name = 'strain'

# Step 3: display
df
```
